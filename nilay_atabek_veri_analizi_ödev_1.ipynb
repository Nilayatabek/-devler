```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import statsmodels.api as sm
import numpy as np
```

```python
da = pd.read_csv("winequality-white (1).csv")
```

```python
da.head()
```

```python
da.rename(columns={'fixed acidity': 'fixed_acidity'}, inplace=True)
da.rename(columns={'volatile acidity': 'volatile_acidity'}, inplace=True)
da.rename(columns={'citric acid': 'citric_acid'}, inplace=True)
da.rename(columns={'residual sugar': 'residual_sugar'}, inplace=True)
da.rename(columns={'free sulfur dioxide': 'free_sulfur_dioxide'}, inplace=True)
da.rename(columns={'total sulfur dioxide': 'total_sulfur_dioxide'}, inplace=True)
model = sm.OLS.from_formula("quality ~ fixed_acidity+volatile_acidity+citric_acid+residual_sugar+free_sulfur_dioxide+total_sulfur_dioxide+fixed_acidity+density+chlorides+pH+sulphates+alcohol" , data=da)
result = model.fit()
result.summary()
```

Çıkarımlarım Şu şekilde:
P>|t| olduğu takdirde değişkenin modeldeki payı anlamlı düzeyde değildir. Citric acid, total sülfür dioksid, chlorides değişkenleri bu lineer regresyon denklemine alınmamalıdır. Diğerleri 0 çıktığından dolayı modele alınması gereken değişkenlerdir. Density kavramının da standart error'u çok yüksek olduğu için modelin doğru tahminlemesini ve kalitesini düşürecektir. Bunun dışında coef başlığı altındaki değerler de denklemde gözükecek katsayılardır.
Bu standart sapması çok büyük ve anlamlılık düzeyi az olan katsayılı değişkenleri çıkardığımda durum şuna dönüşecek:

```python
da.rename(columns={'fixed acidity': 'fixed_acidity'}, inplace=True)
da.rename(columns={'volatile acidity': 'volatile_acidity'}, inplace=True)
da.rename(columns={'citric acid': 'citric_acid'}, inplace=True)
da.rename(columns={'residual sugar': 'residual_sugar'}, inplace=True)
da.rename(columns={'free sulfur dioxide': 'free_sulfur_dioxide'}, inplace=True)
da.rename(columns={'total sulfur dioxide': 'total_sulfur_dioxide'}, inplace=True)
model = sm.OLS.from_formula("quality ~ fixed_acidity+volatile_acidity+residual_sugar+free_sulfur_dioxide+fixed_acidity+pH+sulphates+alcohol" , data=da)
result = model.fit()
result.summary()
```
